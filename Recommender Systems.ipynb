```python
import numpy as np
import pandas as pd
```

```python
column_names = ["id_user","item_id","rate","timestamp"]
```

```python
df = pd.read_csv("u.data.csv",names = column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie.csv")
```

```python
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on = "item_id")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("white")
```

```python
%matplotlib inline
```

```python
df.groupby("title")["rate"].mean().sort_values(ascending = False).head()
```

```python
df.groupby("title")["rate"].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(df.groupby("title")["rate"].mean())
```

```python
ratings.head()
```

```python
ratings["num_of_ratings"] = pd.DataFrame(df.groupby("title")["rate"].count())
ratings.head()
```

```python
ratings["num_of_ratings"].hist(bins=70)
```

```python
ratings["rate"].hist(bins = 70)
```

```python
sns.jointplot(x = "rate", y ="num_of_ratings",data = ratings)
```

```python
moviemat = df.pivot_table(index = "id_user",columns = "title",values = "rate")
moviemat.head()
```

```python
ratings.sort_values("num_of_ratings",ascending = False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat["Star Wars (1977)"]
liarliar_user_ratings = moviemat["Liar Liar (1997)"]
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
similar_to_starwars.head(10)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns = ["Correlations"])
corr_starwars.dropna(inplace = True)
corr_starwars.head()
```

```python
corr_starwars.sort_values("Correlations",ascending = False).head(10)
#In the below shown head we can see that movies with low no of ratings are at the top so inorder to avoid this we have to filter the movies according to the no of ratings.
#So refer to the previous histogram and select a threshold for the no of ratings!!
```

```python
corr_starwars = corr_starwars.join(ratings["num_of_ratings"])
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars["num_of_ratings"] > 100].sort_values("Correlations",ascending = False).head(10)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns = ["Correlations"])
corr_liarliar.dropna(inplace = True)
corr_liarliar = corr_liarliar.join(ratings["num_of_ratings"])
corr_liarliar[corr_liarliar["num_of_ratings"] > 100].sort_values("Correlations",ascending = False).head(10)                                          
```
